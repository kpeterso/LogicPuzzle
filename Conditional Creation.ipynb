```python
dimensions = 2
variables = 2
```

```python
#create integer value for each cell in grid
grid = {}
#i is row
for r in range(variables):
    grid[r] = {}
    #c is column
    for c in range(variables):
        grid[r][c] = c+(2*r)+1
```

```python
#go through each row and column
#at an individual cell, go through rows, and then columns
constraintList = []
for r in range(variables):
    #c is column
    for c in range(variables):
        val = grid[r][c]
        e = [val]
        for i in range(variables):
            if grid[i][c] != val:
                constraintList.append([-val, -grid[i][c]])
                e.append(grid[i][c])
        for j in range(variables):
            if grid[r][j] != val:
                constraintList.append([-val, -grid[r][j]])
                e.append(grid[r][j])
        constraintList.append(e)
```

```python
constraintList
```

```python
constraintDict = {}
constraintDict[dimensions] = {}
print(len(constraintList))
newList = []
for i in constraintList:
    if i not in newList:
        newList.append(i)
        
print(len(newList))
constraintDict[dimensions][variables] = newList
```

```python
dimensions = 3
variables = 2
#create integer value for each cell in grid
grid = {}
#i is row
for r in range(variables):
    grid[r] = {}
    #c is column
    for c in range(variables):
        grid[r][c] = {}
        for a in range(variables):
            grid[r][c][a] = c+(variables*r)+(variables**2)*a+1
```

```python
#go through each row and column
#at an individual cell, go through rows, and then columns
constraintList = []
for r in range(variables):
    #c is column
    for c in range(variables):
        for a in range(variables):
            val = grid[r][c][a]
            e = [val]
            #go through each row and col, keeping arm the same
            for i in range(variables):
                for j in range(variables):
                    if grid[i][j][a] != val:
                        constraintList.append([-val, -grid[i][j][a]])
                        e.append(grid[i][j][a])
            for i in range(variables):
                for k in range(variables):
                    if grid[i][c][k] != val:
                        constraintList.append([-val, -grid[i][c][k]])
                        e.append(grid[i][c][k])
            for j in range(variables):
                for k in range(variables):
                    if grid[r][j][k] != val:
                        constraintList.append([-val, -grid[r][j][k]])
                        e.append(grid[r][j][k])
            constraintList.append(e)
```

```python
print(len(constraintList))
newList = []
for i in constraintList:
    if i not in newList:
        newList.append(i)
        
print(len(newList))
```
